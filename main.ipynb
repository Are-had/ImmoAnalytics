```python

files = [
    ("DATA/ValeursFoncieres-2020-S2.txt", 2020),
    ("DATA/ValeursFoncieres-2021.txt", 2021),
    ("DATA/ValeursFoncieres-2022.txt", 2022),
    ("DATA/ValeursFoncieres-2023.txt", 2023),
    ("DATA/ValeursFoncieres-2024.txt", 2024),
    ("DATA/ValeursFoncieres-2025-S1.txt", 2025)
]

dfs = []
for file, annee in files:
    temp_df = pd.read_csv(file, sep="|", low_memory=False)
    temp_df["Annee"] = annee
    dfs.append(temp_df)

df_total = pd.concat(dfs, ignore_index=True)
df_total.to_csv("DATA/ValeursFoncieres_all.csv", index=False)
print(f"Total: {len(df_total)} lignes")
df_total.head()
```
